```python
#matrix multiplication

#rules: 1. inner dimensions must match
#       2. the resulting shape is the outer dimensions


tensor1 = torch.rand(2, 3) # first dimension is the number of rows, second dimension is the number of columns
tensor2 = torch.rand(3, 4)
tensor3 = torch.rand(4, 5)

print(tensor1 @ tensor2) #shape (2, 4)
print("----")
print(tensor2 @ tensor3) #shape (3, 5)
print("----")
print(tensor1 @ tensor2 @ tensor3) #shape (2, 5) because the inner dimensions (4) match and the resulting shape is the outer dimensions (2, 5)
```

```python
# transpose

tensor_1 = torch.rand(3, 4)
print(tensor_1)
tensor_2 = torch.rand(3, 4) # this cannot be matrix multiplied with tensor_1 because the inner dimensions do not match
print(tensor_2)

tensor_2 = tensor_2.T # transpose of tensor_2, now it has shape (4, 3) and can be matrix multiplied with tensor_1

print(tensor_2)
print(tensor_1 @ tensor_2) # shape (3, 3) because the inner dimensions (4) match and the resulting shape is the outer dimensions (3, 3)
```

```python
x = torch.arange(1, 10)
print(x, x.shape)
z = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(z, z.shape)
# add a dimension to x
x_reshaped = x.reshape(1, 9)
print(x_reshaped, x_reshaped.shape)
x_view = x.view(1, 9) # view is similar to reshape but it does not copy the data, it just changes the shape of the tensor
print(x_view, x_view.shape)
```

```python
# stacking tensors
x_stacked = torch.stack([x, x, x], dim=0) # stack along the first dimension
print(x_stacked, x_stacked.shape)
x_stacked = torch.stack([x, x, x], dim=1) # stack along the second dimension
print(x_stacked, x_stacked.shape)
print(x.reshape(9, 1))
```
