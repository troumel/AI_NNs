## Forward Pass

```python
# ===============forward pass================

#batch of data
N = 10

#each data point has one input feature and one output feature
D_in = 1
D_out = 1

#create input data
x = torch.randn(N, D_in)

#create true targets and add some noise
true_w = torch.tensor([[2.0]])
true_b = torch.tensor([[1.0]])
y_true = x @ true_w + true_b + 0.1 * torch.randn(N, D_out)
#print(y_true)

#initialize random weights
w = torch.randn(D_in, D_out, requires_grad=True)
b = torch.randn(D_out, requires_grad=True)

print(w)
print(b)

y_hat = x @ w + b
print(f"predicted y: {y_hat} \ntrue y: {y_true}")
```

```python
# mean squared error loss
loss = torch.mean((y_hat - y_true) ** 2)
print(f"loss: {loss}") # this is the value we want to minimize

loss.backward()

print(f"gradient of w: {w.grad}")
print(f"gradient of b: {b.grad}")
```

```python
learning_rate = 0.01
epochs = 100

w = torch.randn(D_in, D_out, requires_grad=True)

b = torch.randn(D_out, requires_grad=True)
for epoch in range(epochs):
    # forward pass
    y_hat = x @ w + b
    loss = torch.mean((y_hat - y_true) ** 2)

    # backward pass
    loss.backward()

    # update weights using gradient descent
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

        # zero the gradients after updating
        w.grad.zero_()
        b.grad.zero_()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```
